# Filtered FHV (Run One time for CSV)

```python
#First Filter
#yd_filtered_locations = yd_filtered_locations.drop(['SR_Flag'], axis = 1)
yd_filtered_locations.tpep_pickup_datetime = pd.to_datetime(yd_filtered_locations.tpep_pickup_datetime)
yd_filtered_locations.tpep_dropoff_datetime = pd.to_datetime(yd_filtered_locations.tpep_dropoff_datetime)
yd_filtered_locations['time_of_ride'] = yd_filtered_locations.tpep_dropoff_datetime - yd_filtered_locations.tpep_pickup_datetime
#fhv_data['pickup_month'] = fhv_data.pickup_datetime.dt.month
yd_filtered_locations['tpep_pickup_day'] = yd_filtered_locations.tpep_pickup_datetime.dt.day
yd_filtered_locations['tpep_pickup_hour'] = yd_filtered_locations.tpep_pickup_datetime.dt.hour
yd_filtered_locations['tpep_dropoff_day'] = yd_filtered_locations.tpep_dropoff_datetime.dt.day
yd_filtered_locations['tpep_dropoff_hour'] = yd_filtered_locations.tpep_dropoff_datetime.dt.hour
time_of_ride_seconds = []
for i in yd_filtered_locations['time_of_ride']:
    time_of_ride_seconds.append(td.total_seconds(i)/60)
yd_filtered_locations['time_of_ride_minutes'] = time_of_ride_seconds

#Second Filter
#print(fhv_data_filtered_locations.pickup_datetime)
yd_filtered_locations.tpep_pickup_datetime = pd.to_datetime(yd_filtered_locations.tpep_pickup_datetime)
yd_filtered_locations['tpep_pickup_month'] = yd_filtered_locations['tpep_pickup_datetime'].dt.strftime('%m')

#Append the Day, Number of the Day for the Week, Month, and Season
import calendar
yd_filtered_locations.tpep_pickup_datetime = pd.to_datetime(yd_filtered_locations.tpep_pickup_datetime)
fhv_weekday = []
fhv_weekday_number = []
fhv_month = []

for i in yd_filtered_locations.tpep_pickup_datetime:
    weekday = calendar.weekday(i.year, i.month, i.day)
    
    #fhv_month.append(i.month)
    if weekday == 0:
        fhv_weekday.append("Monday")
        fhv_weekday_number.append(0)
    elif weekday == 1:
        fhv_weekday.append("Tuesday")
        fhv_weekday_number.append(1)
    elif weekday == 2:
        fhv_weekday.append("Wednesday")
        fhv_weekday_number.append(2)
    elif weekday == 3:
        fhv_weekday.append("Thursday")
        fhv_weekday_number.append(3)
    elif weekday == 4:
        fhv_weekday.append("Friday")
        fhv_weekday_number.append(4)
    elif weekday == 5:
        fhv_weekday.append("Saturday")
        fhv_weekday_number.append(5)
    elif weekday == 6:
        fhv_weekday.append("Sunday")
        fhv_weekday_number.append(6)
yd_filtered_locations['pickup_weekday'] = fhv_weekday
yd_filtered_locations['pickup_weekday_num'] = fhv_weekday_number

#Determine the Seasons
seasons = []
seasons_number = []
for month in yd_filtered_locations.tpep_pickup_month:
    month = int(month)
    #December to February: Winter | March to May: Spring | June to July: Summer | August to October: Fall
    if month == 12 or month == 1 or month == 2:
        seasons.append("Winter")
        seasons_number.append(1)
    elif month >= 3 and month <= 5:
        seasons.append("Spring")
        seasons_number.append(2)
    elif month >= 6 and month <= 8:
        seasons.append("Spring")
        seasons_number.append(3)
    elif month >= 9 and month <= 11:
        seasons.append("Spring")
        seasons_number.append(4)
        
yd_filtered_locations['seasons'] = seasons
yd_filtered_locations['seasons_number'] = seasons_number

# Length of Ride Interval Categories (LORI)
# 1:'0-5' | 2:'5-10' | 3:'10-15' | 4:'15-20' | 5:'20-30' | 6:'30-45' | 7:'45+'
LORI = []
SNL = []

#Create a column for LORI
for time in yd_filtered_locations.time_of_ride_minutes:
    time = round(time, 0)
    if time <= 5:
        LORI.append(1)
    elif time <= 10:
        LORI.append(2)
    elif time <= 15:
        LORI.append(3)
    elif time <= 20:
        LORI.append(4)
    elif time <= 30:
        LORI.append(5)
    elif time <= 45:
        LORI.append(6)
    else:
        LORI.append(7)

for time in yd_filtered_locations.time_of_ride_minutes:
    time = round(time, 0)
    if time <= 20:
        SNL.append(0)
    elif time >= 20:
        SNL.append(1)
        
yd_filtered_locations['LORI'] = LORI

#First, Second, & Third Shift
typeOfShiftArray = []

#Categorize: First Shift from 7 AM to 3 PM, Second Shift from 3 PM to 11 AM, & Third Shift from 11 AM to 7 AM.
for time in yd_filtered_locations.tpep_pickup_hour:
    if time >= 7 and time <= 15:
        typeOfShiftArray.append(1)
    elif time <= 23 and time >= 15:
        typeOfShiftArray.append(2)
    elif time <= 7:
        typeOfShiftArray.append(3)
        
yd_filtered_locations['ShiftType'] = typeOfShiftArray

#Day, Afternoon, Evening, or Night (DAEN)
#5 AM to 11 AM, 12PM to 4PM, 5PM to 8PM, 9PM to 4 AM

DAEN = []

for time in yd_filtered_locations.tpep_pickup_hour:
    if time >= 4 and time <= 10:
        DAEN.append(1)
    elif time <= 16 and time >= 11:
        DAEN.append(2)
    else:
        DAEN.append(3)

yd_filtered_locations['DAEN'] = DAEN
```
